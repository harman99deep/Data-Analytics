## Part of Speech (POS) Tagging and Dependency Parsing

```python
# Assigning lemmatized text to Variables
x = lemm_join  # Digital Music 
y = lemm2_join # Musical Instruments
```

Digital Music

```python
x[0]
```

```python
# Noun Chunks
#x= combined_data['reviewText']
contents = x
text = str(contents)
doc = nlp(text)
print('Input Text:\n',doc,'\n')
for chunk in doc.noun_chunks:
   print('"Text"=',chunk.text,'\t"Root text"=',chunk.root.text, '\t"Root dep"=',chunk.root.dep,
          '\t"Root head text"=',chunk.root.head.text)
```

```python
# Navigating Parsing Tree
contents = x                                                             
text = str(contents)                                                              
doc = nlp(text)                                                 # Input text
for token in doc:
    print('Text=',token.text,', "\tDEP"=',token.dep_, ', "\tHead text"=',token.head.text, ', "\tHead POS"=',token.head.pos_,', "\tChildren"=',
            [child for child in token.children])
```

#### Text: The original token text.
#### Dep: The syntactic relation connecting the child to the head. Terms are given in learning consolidation.
#### Head text: The original text of the token head.
#### Head POS: The part-of-speech tag of the token head.
#### Children: The immediate syntactic dependents of the token.

```python
from spacy import displacy                                   # Import displacy for Visualizing Dependencies
doc = nlp(x[0])                            # Input text
print('Input Text:\n\n',doc)                                   # Print input
displacy.render(doc, style="dep", jupyter= True)
```

```python
# Create an nlp object
doc = nlp(x[0])
 
# Iterate over the tokens
for token in doc:
    # Print the token and its part-of-speech tag
    print(token, ':', token.tag_,',', token.pos_,',', spacy.explain(token.tag_))
```

```python
#Figuring out the distribution of Digital Music
import matplotlib.pyplot as plt

print('Ratings Count')
print(dm_df['ratings'].value_counts())
print('\n')
dm_df['ratings'].value_counts().plot(kind = 'bar')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.title('Overall Ratings Count')

```

```python
def f(row):
    
   if row['ratings'] == 3.0:
        val = 'Neutral'
   elif row['ratings'] == 1.0 or row['ratings'] == 2.0:
        val = 'Negative'
   elif row['ratings'] == 4.0 or row['ratings'] == 5.0:
        val = 'Positive'
   else:
        val = -1
   return val
```

```python
dm_df['sentiment'] = dm_df.apply(f, axis=1)
dm_df.head()
```

```python
dm_df['sentimentScore'] = dm_df['sentiment']
dm_df['sentimentScore'][dm_df.sentimentScore == 'Positive'] = 2
dm_df['sentimentScore'][dm_df.sentimentScore == 'Neutral'] = 1
dm_df['sentimentScore'][dm_df.sentimentScore == 'Negative'] = 0
```

```python
dm_df['ratings']= dm_df['ratings'].astype(int)
```

```python
dm_df.head()
```

```python
# Vectorize training data using TfidfVectorizer
# Remark: TfidfVectorizer = CountVectorizer+TfidfTransformer
nltk.download('stopwords')
from nltk.corpus import stopwords
vectorizer = TfidfVectorizer(ngram_range=(1,1),max_features=5000,stop_words=stopwords.words('english'))  
```

```python
tf_x_train = vectorizer.fit_transform(X_train)
tf_x_test = vectorizer.transform(X_test)
```
